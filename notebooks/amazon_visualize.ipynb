# Amazon Data Visualization

```python
'''
This notebook will explore visualizations associated to the preprocessed data from amazon_data.ipynb
It will explore differentt distributions, and correlations associated to various potential features
identified.
'''

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set()
```

```python
# constants
base_path = os.path.expanduser('~') + '/'
data_path = 'Downloads/video_game_data_preprocessed.csv'
```

## Import Data

```python
d = pd.read_csv(base_path + data_path, low_memory = False)
```

## Distribution Visualization

```python
def plot_hist(d, column, title, xlabel, ylabel = 'Counts'):
    '''
    This funuction will plot a histogram (distribution) associated to a particular column
    identified
    '''
    
    plt.clf()
    plt.hist(d[column].dropna(), alpha = 0.5, bins = 100)
    plt.xlabel(xlabel)
    plt.xticks(rotation = 45)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
plot_hist(
    d,
    'description_count',
    'Description Word Count Distribution',
    'Description Count'
)
```

```python
plot_hist(
    d,
    'title_wc',
    'Title Word Count Distribution', 
    'Title Word Count'
)
```

```python
plot_hist(
    d, 
    'summary_wc',
    'Summary Word Count Frequency Distribution', 
    'Summary Word Count'
)
```

```python
plot_hist(
    d,
    'review_wc',
    'Review Count Frequency Distribution',
    'Review Word Count'
)
```

```python
plot_hist(
    d, 
    'recommended_item_counts', 
    'Frequency of Recommended Items Count',
    'Recommended Item Count'
)
```

```python
plot_hist(
    d, 
    'review_year',
    'Frequency of Review Years',
    'Review Count per Year'
)
```

```python
avg_votes_yearly = d.groupby('review_year')['vote'].agg(['mean', 'sum']).reset_index()
```

```python
avg_votes_yearly.head()
```

```python
def plot_bar(d, xcol, ycol, xlabel, ylabel, title):
    '''
    This function will plot a bar graph given the input params
    '''
    plt.clf()
    plt.bar(data = d, x = xcol, height = ycol, alpha = 0.65)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
plot_bar(
    d = avg_votes_yearly, 
    xcol = 'review_year', 
    ycol = 'mean', 
    xlabel = 'Year',
    ylabel = 'Avg Votes', 
    title = 'Average Yearly Votes'
)
```

```python
plot_bar(
    d = avg_votes_yearly, 
    xcol = 'review_year', 
    ycol = 'sum', 
    xlabel = 'Year',
    ylabel = 'Total Votes', 
    title = 'Total Yearly Votes'
)
```

## Correlation Visualization

```python
def plot_scatter(d, xcol, ycol, xlabel, ylabel, title):
    '''
    This function will plot the xcolumns against the y columns to check for 
    correlation
    '''
    
    plt.clf()
    plt.scatter(d[xcol], d[ycol], alpha = 0.5)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
plot_scatter(
    d,
    'title_wc',
    'summary_wc',
    'Title Word Count',
    'Summary Word Count',
    'Title vs Summary Word Count'
)
```

```python
plot_scatter(
    d,
    'review_wc',
    'summary_wc',
    'Review Word Count',
    'Summary Word Count',
    'Review vs Summary Word Count'
)
```

```python
plot_scatter(
    d,
    'review_wc',
    'title_wc',
    'Review Word Count',
    'Title Word Count',
    'Review vs Title Word Count'
)
```

```python
plot_scatter(
    d,
    'review_wc',
    'vote',
    'Review Word Count',
    'Votes',
    'Review vs Votes'
)
```

```python
plot_scatter(
    d,
    'summary_wc',
    'vote',
    'Summary Word Count',
    'Votes',
    'Summary vs Votes'
)
```

```python
plot_scatter(
    d,
    'description_count',
    'vote',
    'Description Word Count',
    'Votes',
    'Description vs Votes'
)
```
